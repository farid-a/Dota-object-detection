```python
# foldery z danymi
folder_with_data = '/content/drive/MyDrive/DOTA/Dataset'

train_images_dir = folder_with_data + '/Train/mod/images'
train_annotations_dir = folder_with_data + '/Train/mod/annots'
```

```python
# stworzenie list X - dane (obrazy) y - adnotacje
def create_dataset(all_images_folder, all_annots_folder, folder_number):
  X = []
  y = []
  all_images = glob.glob(all_images_folder + f'{folder_number}/*.png')
  all_annots = glob.glob(all_annots_folder + f'{folder_number}/*.txt')
  print(len(all_images))
  print(len(all_annots))

  for annot in all_annots:
    annot_file_name = os.path.basename(annot).replace('.txt', '')
    for image in all_images:
      image_file_name = os.path.basename(image).replace('.png', '')
      if annot_file_name == image_file_name:
        X.append(image)
        y.append(annot)
        break
  return X, y
```

```python
X = []
y = []

for i in range(14):
  X_temp, y_temp = create_dataset(train_images_dir, train_annotations_dir, folder_number=i)
  X += X_temp
  y += y_temp
```

```python
classes = {
    'BG': -1,
    'plane': 0,
    'ship': 1,
    'storage-tank': 2,
    'baseball-diamond': 3,
    'tennis-court': 4,
    'basketball-court': 5,
    'ground-track-field': 6,
    'harbor': 7,
    'bridge': 8,
    'large-vehicle': 9,
    'small-vehicle': 10,
    'helicopter': 11,
    'roundabout': 12,
    'soccer-ball-field': 13,
    'swimming-pool': 14,
    'container-crane': 15
}
```

```python
new_data = '/content/drive/MyDrive/DOTA/Dataset/dotayolo'
```

```python
def return_folder_number(full_path):
  path_ = os.path.dirname(full_path)
  curr_folder = os.path.basename(path_)
  return ''.join(re.findall('[0-9]+', curr_folder))
```

```python
# algorym do transformacji adnotacji do formatu YOLO
def change_annots_to_YOLO(X, y, number_of_dirs=14):
  number_of_files = 0
  # for dir_num in range(0, number_of_dirs):
  #  os.mkdir(f"{new_data}/images/{str(dir_num)}") 
  #  os.mkdir(f"{new_data}/labels/{str(dir_num)}")
  
  for image_name_fullpath, annot_name_fullpath in zip(X, y):
    curr_dir_number_img = return_folder_number(image_name_fullpath)
    curr_dir_number_ann = return_folder_number(annot_name_fullpath)

    assert curr_dir_number_img == curr_dir_number_ann 
    curr_dir_number = curr_dir_number_img 

    image_name = os.path.basename(image_name_fullpath)
    annot_name = os.path.basename(annot_name_fullpath)

    w, h = imagesize.get(image_name_fullpath)

    shutil.copy(image_name_fullpath, f"{new_data}/images/{curr_dir_number}/{image_name}")
    new_annot = open(f"{new_data}/labels/{curr_dir_number}/{annot_name}", 'w')
    
    all_lines = open(annot_name_fullpath).readlines()
    
    number_of_files += 1
    print('obraz numer:', number_of_files)
    
    line_to_write = ''
    # bounding boxy
    if len(all_lines) > 0:
      if len(all_lines[0].split(' ')) == 10:
        first_line = 0
      else:
        first_line = 2

      for line in all_lines[first_line:]:
        line = line.replace('\n', '')
        assert len(line.split(' ')) == 10
        x1, y1, x2, y2, x3, y3, x4, y4, class_name, diff = line.split(' ')

        xs = [float(x1), float(x2), float(x3), float(x4)]
        ys = [float(y1), float(y2), float(y3), float(y4)]
        assert xs.count(max(xs)) == xs.count(min(xs)) and ys.count(max(ys)) == ys.count(min(ys))

        bb_w = max(xs) - min(xs)
        bb_h = max(ys) - min(ys)
        x_mid_norm = (max(xs) - bb_w / 2.0) / w
        y_mid_norm = (max(ys) - bb_h / 2.0) / h
        obj_w_norm = bb_w / w
        obj_h_norm = bb_h / h

        line_to_write += f'{str(classes[class_name])}' + " " + f'{str(x_mid_norm)}' + " " + f'{str(y_mid_norm)}' + " " + f'{str(obj_w_norm)}' + " " + f'{str(obj_h_norm)}\n'
      n = new_annot.write(line_to_write)
    new_annot.close()
```
