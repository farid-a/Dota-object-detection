```python
map_classes = {
  0: 'plane',
  1: 'ship',
  2: 'storage-tank',
  3: 'baseball-diamond',
  4: 'tennis-court',
  5: 'basketball-court',
  6: 'ground-track-field',
  7: 'harbor',
  8: 'bridge',
  9: 'large-vehicle',
  10: 'small-vehicle',
  11: 'helicopter',
  12: 'roundabout',
  13: 'soccer-ball-field',
  14: 'swimming-pool',
  15: 'container-crane'
}
```

```python
# zapis wynik√≥w do zadanego formatu
def save_results(tensor, img_name, folder_num):
    bb_file_name = os.path.basename(img_name).replace('.png', '.txt')
    with open(f'/content/drive/MyDrive/DOTA/Dataset/results_yolov5_mid/{folder_num}/{bb_file_name}', 'w') as file:
      for c in tensor:
        file.write(f'{map_classes[int(c[5])]} {c[4]} {int(c[0])} {int(c[1])} {int(c[2])} {int(c[3])}\n')
```

```python
# model
model = torch.hub.load('/content/drive/MyDrive/DOTA/Dataset/yolov5', 
                       'custom', 
                       path='/content/drive/MyDrive/DOTA/Dataset/yolov5/runs/train/dota-yolov5m2/weights/last.pt', 
                       source='local'
                       ) 

model.conf = 0.468  # minimalna warto≈õƒá pewno≈õci dla algorytmu NMS
model.iou = 0.6  # warto≈õƒá progowa IoU dla algorytmu NMS
model.multi_label = True  # wiele obiekt√≥w w jednym wycinku obrazu
model.max_det = 1600  # maksymalna ilo≈õƒá instancji dla danego obrazu
```

```python
# numer folderu
folder_num = 5

imgs = glob.glob(f'/content/drive/MyDrive/DOTA/Dataset/Val/mod/images{folder_num}/*.png')
print(imgs)
print(len(imgs))

for img in imgs:
  # detekcja
  results = model(cv2.imread(img)[..., ::-1], size=928)
  # zapis
  save_results(results.xyxy[0], img, folder_num)

# sprawdzenie poprawno≈õci zapisu
print(len(glob.glob(f'/content/drive/MyDrive/DOTA/Dataset/results_yolov5_mid/{folder_num}/*.txt')))
```
