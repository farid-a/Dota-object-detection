```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 0.2)
```

```python
plt.plot(x, np.heaviside(x, 0))
plt.xlabel('h')
plt.ylabel(r'$\Phi(h)$')
plt.title('Funkcja skokowa')
```

```python
plt.plot(x, tf.math.sigmoid(x))
plt.xlabel('h')
plt.ylabel(r'$\Phi(h)$')
plt.title('Funckja sigmoidalna')
```

```python
x = np.arange(-10., 10., 0.2)
plt.plot(x, tf.nn.relu(x))
plt.xlabel('h')
plt.ylabel(r'$\Phi(h)$')
plt.title('Funkcja ReLU')
```

```python
x = np.arange(-10., 10., 0.2)
plt.plot(x, tf.math.tanh(x))
plt.xlabel('h')
plt.ylabel(r'$\Phi(h)$')
plt.title('Funkcja tangens hiperboliczny')
```
